```python
from nltk.tokenize import RegexpTokenizer

text = "Com um passe de Eli Manning para Plaxico Burress a 39 segundos do fim, o New York Giants anotou o touchdown decisivo e derrubou o favorito New England Patriots por 17 a 14 neste domingo, em Glendale, no Super Bowl XLII."
tokenizer = RegexpTokenizer(r'\w+') #\w returns string com characters from a to Z, digits 0-9, _
                                    # + one or more occurences
                                    # pythex.org
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
# without numbers
from nltk.tokenize import RegexpTokenizer
# [] - specify a set of characters
tokenizer = RegexpTokenizer(r'[a-zA-Z]\w*') # all tokens containing letters; * all occurences
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
# tokens frequency
text = "Com um passe de Eli Manning para Plaxico Burress a 39 segundos do fim, o New York Giants anotou o touchdown decisivo e derrubou o favorito New England Patriots por 17 a 14 neste domingo, em Glendale, no Super Bowl XLII."
tokens = nltk.word_tokenize(text)
freq = nltk.FreqDist(tokens)
print(freq.most_common(10))
```

## Tagging

```python
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger

tokens = nltk.word_tokenize(corpus)

sentences_train = mac_morpho.tagged_sents()

tagger = UnigramTagger(sentences_train)

tagged = tagger.tag(tokens)

print(tagged)
```

```python
## Adding a standard tag - when tag is None, it considers as substantive
from nltk.corpus import mac_morpho
from nltk.tag import UnigramTagger
from nltk.tag import DefaultTagger

tokens = nltk.word_tokenize(corpus)

tag_std = DefaultTagger('N') # N = substantive
sentences_train = mac_morpho.tagged_sents()

tagger = UnigramTagger(sentences_train, backoff=tag_std)

tagged = tagger.tag(tokens)

print(tagged)
```

## Chuncking

```python
from nltk.chunk import RegexpParser

pattern = 'NP: {<NPROP><NPROP> | <N><N>}' # named entities
grammatic_analysis = RegexpParser(pattern)

tree = grammatic_analysis.parse(tagged)
print(tree)
```
